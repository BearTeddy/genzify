```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# First URL
url = "https://gabb.com/blog/teen-slang/?srsltid=AfmBOooFDnO8kTACaiZc-Gv2EnBI7n_ZwuZhFSvQHrxDTKw5RgQXtTkb"

response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
```

```python
# Get the text from the article
article = soup.find_all(attrs={"class": "has-medium-font-size"})
print(article)
```

```python
df = pd.DataFrame()
data = []
for line in article:
    key = ' – '
    kv = line.text.split(key)
    data.append(kv)
    
    
df = pd.DataFrame(data, columns=['slang', 'meaning'])    
df.to_csv('slang.csv', index=False)
```
