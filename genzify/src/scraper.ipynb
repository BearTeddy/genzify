```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# First URL
url = "https://gabb.com/blog/teen-slang/?srsltid=AfmBOooFDnO8kTACaiZc-Gv2EnBI7n_ZwuZhFSvQHrxDTKw5RgQXtTkb"

response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
```

```python
# Get the text from the article
article = soup.find_all(attrs={"class": "has-medium-font-size"})
print(article)
```

```python
df = pd.DataFrame()
data = []
for line in article:
    key = ' – '
    kv = line.text.split(key)
    data.append(kv)
    
    
df = pd.DataFrame(data, columns=['slang', 'meaning'])    
df.to_csv('slang.csv', index=False)
```

```python
# Gen Z Bible Area
from selenium import webdriver
driver = webdriver.Chrome()  # Make sure to have the appropriate WebDriver installed
rootLink = "https://genz.bible"

Link = "https://genz.bible/Genesis"
driver.get(Link)
# Get using Xpath
html_content = driver.page_source
soup = BeautifulSoup(html_content, "html.parser")

span_element = soup.findAll('span', class_="MuiTypography-root MuiListItemText-primary MuiTypography-body1 MuiTypography-displayBlock")
title = [span.get_text() for span in span_element]

chapterlist = []

for ti in title:
    driver.get(rootLink + "/" + ti)
    html_content = driver.page_source
    soup = BeautifulSoup(html_content, "html.parser")
    ref = soup.find_all('a', class_="MuiButtonBase-root MuiButton-root MuiButton-text")
    chplist = [chapter['href'] for chapter in ref]
    chapterlist.append(chplist)

print(chapterlist)
print(title)
    
#Save the list to a csv file
data = []
for i in range(len(title)):
    for j in range(len(chapterlist[i])):
        data.append([title[i], chapterlist[i][j]])
df = pd.DataFrame(data, columns=['Book', 'Chapter'])
df.to_csv('bible.csv', index=False)

# Close the browser
driver.quit()
```

```python
driver = webdriver.Chrome() 

pd_data = pd.DataFrame([], columns=['GenZ', 'Norm'])
for chap in chapterlist:
    data = []
    for chap in chap:
        driver.get(rootLink + chap)
        html_content = driver.page_source
        soup = BeautifulSoup(html_content, "html.parser")
        verse_genz = soup.find('div', class_="MuiGrid-root MuiGrid-item MuiGrid-grid-xs-6", style='order: 1;')
        verse_normal = soup.find('div', class_="MuiGrid-root MuiGrid-item MuiGrid-grid-xs-6", style='order: 2;')
        try : 
            data.append([verse_genz.get_text(), verse_normal.get_text()])
        except:
            pass
    pd_data = pd.concat([pd_data, pd.DataFrame(data=data, columns=['GenZ', 'Norm'])], ignore_index=True)
pd_data.to_csv('bible_verses_2.csv', index=False)
driver.quit()
```
